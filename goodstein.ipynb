```
def stein(n,d):
  s=0
  i=0
  while n>0:
    r=n%d
    n=n//d
    s+=r*(d+1)**stein(i,d)
    i+=1
  return s

def gs(n,d):
  return stein(n,d)-1
```

```
t=int(4)
i = 2
while i <= 1000:
  print(t)
  t=gs(t,i)
  i += 1
  if(t==0): 
    print(t)
    break; # added
```
